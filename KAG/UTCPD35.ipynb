# Getting 5 day forecast 

Api structure 

```sh
api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={API key}

```

````python
import requests as rq 

def ci_we_5d():
    
    """Function is for the actual testing here"""
    lat = 51.507351
    lon = -0.127758
    city = "London"
    appid = "2b9604fddc281764bd583c2112ecde46"
    endpoint = "https://api.openweathermap.org/data/2.5/forecast/"
    
    # Construct URL 
    url = f"{endpoint}?lat={lat}&lon={lon}&appid={appid}"
    
    # Runnign the request here
    resp = rq.get(url).json()    

    pprint(resp, expand_all="TRUE")

ci_we_5d()
````

# Doing the above method with a params method

````python
def ci_we_5d():
    
    """Function is for the actual testing here"""
    
    city = "London"
    
    we_pa = {
        "lat" : 51.507351,
        "lon" : -0.127758,        
        "appid" : "2b9604fddc281764bd583c2112ecde46"
    }
    
    endpoint = "https://api.openweathermap.org/data/2.5/forecast/"
    
    
    # Runnign the request here
    resp = rq.get(endpoint, params=we_pa).json()    

    pprint(resp, expand_all="TRUE")

ci_we_5d()
````
